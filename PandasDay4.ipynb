# How do i handle missing values of pandas

```python
import pandas as pd
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports')
```

```python
ufo.tail()
```

```python
ufo.isnull().tail()
```

```python
ufo.notnull().tail () #not null inverse method
```

```python
ufo.isnull().sum() #the numbers are missing values of each column
```

```python
ufo[ufo.City.isnull()]
```

```python
ufo.shape
```

```python
ufo.dropna(how='any').shape
```

```python
ufo.dropna(how='all').shape
```

# How do i select multiple rows and columns

```python
import pandas as pd
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports')
```

```python
ufo.head(3)
```

```python
ufo.loc[0, :]
```

```python
ufo.loc[0:2, :]
```

```python
#columns selection
```

```python
ufo.loc[:, 'City']
```

```python
ufo.loc[:, 'City':'State']
```

```python
#combine rows and columns
```

```python
ufo.loc[0:2, 'City':'State']
```

```python
ufo.head(3).drop('Time', axis=1)
```

```python
ufo[ufo.City=='Ithaca']
```

```python
ufo[ufo.City=='Holyoke'].State
```

```python
ufo.loc[ufo.City=='Ithaca',:]
```

```python
#USING LOC METHOD 
ufo.loc[ufo.City=='Holyoke', 'State']
```

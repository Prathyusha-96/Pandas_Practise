# How do i rename columns 

```python
import pandas as pd
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports')
```

```python
ufo.head()
```

```python
ufo.columns
```

```python
#USING rename dataframe method
ufo.rename(columns={'Colors Reported':'Colors_Reported', 'Shape Reported':'Shape_Reported'}, inplace=True)
```

```python
ufo.columns
```

```python
#using another method
ufo_cols = ['city', 'colors reported', 'shape reported', 'state', 'time']
ufo.columns = ufo_cols
```

```python
ufo.head()
```

```python
#change the column names while reading csv file
ufo = pd.read_csv('http://bit.ly/uforeports', names = ufo_cols, header=0)
```

```python
ufo.head()
```

```python
#we have multiple columns
ufo.columns = ufo.columns.str.replace(' ', '_')
```

```python
ufo.columns
```

```python
ufo.head()
```

# How do i remove columns

```python
import pandas as pd
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports')
```

```python
ufo.head()
```

```python
ufo.shape
```

```python
#remove columns
ufo.drop('City', axis=1, inplace=True)
```

```python
ufo.head()
```

```python
ufo.drop(['State', 'Time'], axis=1, inplace=True)
```

```python
ufo.head()
```

```python
#remove rows
ufo.drop([1,2], axis = 0, inplace=True)
ufo.head()
```

# How do i filter rows of a pandas dataframe by column value

```python
import pandas as pd
```

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.shape
```

```python
type(True)
```

```python
movies.head()
```

```python
booleans = []
for length in movies.duration:
    if length >= 200:
        booleans.append('True')
    else:
        booleans.append('False')
```

```python
booleans[0:5]
```

```python
len(booleans)
```

```python
is_long = pd.Series(booleans)
```

```python
is_long_bool = is_long.astype(bool)
is_long_bool
```

```python
is_long.head()
```

```python
movies['genre']
```

```python
is_long = movies.duration >= 200
is_long.head()
```

```python
movies[is_long]
```

```python
movies[movies.duration >= 200]
```

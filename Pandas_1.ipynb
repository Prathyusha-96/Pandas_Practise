```python
#Create a DataFrame from a list of lists. Name the columns "col1", "col2" and "col3".
```

```python
import pandas as pd
```

```python
data_list = [[1,2,3], [4,5,6]]
df = pd.DataFrame(data_list, columns=['col1', 'col2', 'col3'], dtype='int8')
df
```

```python
df.dtypes
```

```python
#Create a DataFrame from a dictionary.
data_dict = {'col1': [1, 2, 3], 'col2': [4, 5, 6]}
df=pd.DataFrame(data_dict, dtype='int8')

df
```

```python
df.dtypes
```

```python
#Create a DataFrame from a numpy array. Name the columns "col1", "col2" and "col3".
import numpy as np
data_nparray = np.array([[1,2,3], [4,5,6]])
df = pd.DataFrame(data_nparray)
df
```

```python
#List-to-Series Conversion
li = [2, 4, 5, 6, 8, 9]
ds = pd.Series(li)
ds
```

```python
#List-to-Series Conversion with Custom Indexing
li = [2, 4, 5, 6, 8, 9, 10, 11]
ds = pd.Series(li, index=[3, 5, 7, 9, 11, 13, 15, 17])
ds
```

```python
# Date Series Generation
date = pd.date_range(start='05-01-2022', end='05-12-2022')
date
```

```python
#Implementing a function on each and every element of a series
ds = pd.Series([2, 4, 5, 6, 8])
ds
modified_series = ds.apply(lambda x:x/2)
modified_series
```

```python
# Dictionary-to-Dataframe Conversion
dict = {'name': ['vinay', 'john', 'deo'], 'age': [25, 26, 20], 'study':['arts', 'civil', 'cs']}
df = pd.DataFrame(dict)
df
```

```python
#2D List-to-Dataframe Conversion
list1 = [[1, 2, 3, 4], [5, 8, 9, 12]]
df = pd.DataFrame(list1)
df
```
